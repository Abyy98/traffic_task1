# Tasks

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
df = pd.read_csv('/content/Task (2) Dataset.csv')
df
```

## 1. Exploratory Data Analysis (EDA)

### Perform summary statistics on the dataset.

```python
df.describe()
```

### Identify and analyze patterns in the data.

```python

numerical_df = df.select_dtypes(include=['number'])
correlation_matrix = numerical_df.corr()
correlation_matrix
```

### Visualize the distribution of key variables (e.g., Traffic_Index, Date).

```python
sns.displot(data= df, x='TrafficIndexLive', height=6, aspect=2)
plt.title(' Traffic Index Live')
plt.xlabel(' TrafficIndexLive')
plt.show()
```

```python
sns.displot(data= df, x='Datetime', height=6, aspect=2)
plt.title('Date time')
plt.xlabel('Datetime')
plt.show()

```

```python
df.columns
```

### Explore relationships between variables (e.g., Traffic_Index vs. Weather_Condition).


```python
sns.scatterplot(data= df, x='TrafficIndexLive', y='City')
sns.displot(data= df, x='TrafficIndexLive', height=6, aspect=2)
plt.title('Relationship Between TrafficIndexLive and city')
plt.xlabel('TrafficIndexLive')
plt.ylabel('City')
plt.show()
```

```python
sns.scatterplot(data= df, x='TrafficIndexLive', y='TravelTimeHistoric')
sns.displot(data= df, x='TrafficIndexLive', height=6, aspect=2)
plt.title('Relationship Between TrafficIndexLive and TravelTimeHistoric')
plt.xlabel('TrafficIndexLive')
plt.ylabel('TravelTimeHistoric')
plt.show()
```
